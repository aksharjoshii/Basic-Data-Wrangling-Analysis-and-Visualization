###  28. Import the dataset occupation.csv and assign object as users

```python
users = pd.read_csv('data/occupation.csv', sep='|')
users.head()
```

### 29. Discover what is the mean age per occupation

```python
occupation_mean_age = users.groupby('occupation')['age'].mean()
occupation_mean_age.to_frame()
```

### 30. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['gender_enc'] = np.where(users['gender']== 'M', 1,0)
occupation_gender = users.groupby(['occupation', 'gender_enc']).size().unstack(fill_value=0)
occupation_gender['Male Ratio'] = occupation_gender[1] / (occupation_gender[1] + occupation_gender[0])
occupation_gender = occupation_gender.sort_values('Male Ratio', ascending=False)

occupation_gender.rename(columns={0 : 'F', 1 : 'M'}, inplace=True)
occupation_gender
```

### 31. For each occupation, calculate the minimum and maximum ages

```python
occupation_age = users.groupby('occupation')['age'].agg(['min', 'max'])
occupation_age.rename(columns={'min': 'min_age', 'max' :'max_age'}, inplace=True)
```

```python
occupation_age
```

### 32. For each combination of occupation and gender, calculate the mean age

```python
oc_gen = users.groupby(['occupation', 'gender'])['age'].apply('mean')
oc_gen = oc_gen.to_frame()
oc_gen.rename(columns={'age': 'avg age'}, inplace=True)
oc_gen
```

### 33.  For each occupation present the percentage of women and men

```python
oc_gen_perc = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
oc_gen_perc['total'] = oc_gen_perc.sum(axis=1)
oc_gen_perc['Male %'] = (oc_gen_perc['M'] / oc_gen_perc['total']) * 100
oc_gen_perc['Female %'] = (oc_gen_perc['F'] / oc_gen_perc['total']) * 100
oc_gen_perc
```

### 88. Calculate total sales data (quantity) for 2019 for each product and show it using a Pie chart. Understand percentage contribution from each product

```python
products = ['FaceCream', 'FaceWash', 'ToothPaste', 'Soap', 'Shampo',
            'Moisturizer']
Product_Qty_total = []
#lets calculate total quantity of each product in the year 2019
for product in products:
    Product_Qty_total.append(company_data[product].sum())

# lets plot a pie chart 
plt.figure(figsize=(8,7))
plt.pie(Product_Qty_total, 
        labels= products,
        autopct='%1.1f%%'
        )
plt.title('Product sales Analysis', color = 'olive')
plt.legend(fontsize=8, loc = 'lower right')
plt.tight_layout()
plt.show()
```
